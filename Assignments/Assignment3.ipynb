### Instrumental and Experimental Broadening

In class we briefly discussed the Scherrer equation for peak broadening,  
<font size=5> $ \tau = \frac{K \lambda}{\beta cos \theta} $  
<font size=3> Where $\tau$ is the mean size of nanocrystallites. For the following questions, assume that the X-ray wavelength is $1.54 Å$.  

<font color='red'>12. $\beta$, the FWHM of the peak, must be corrected for instrumental broadening before being used. Instrumental broadening is caused by spread in X-ray wavelength (bandwidth). Assume that the bandwidth of your source is 2 eV. What is the broadening in q caused by this bandwidth?

<font color='red'> **Known equations:**

<font color='red'> **$ E = \frac{hc }{\lambda} $**

<font color='red'> **$\Delta \lambda = \frac{\Delta E * \lambda^2}{hc} $**

<font color='red'> **$q = \frac{4π }{\lambda} sinθ$**

<font color='red'> **$\Delta q = \frac{4π }{\lambda^2} sinθ*\Delta \lambda $**

````python
# Constants
wl = 1.54 * 1e-10  # X-ray wavelength given
delta_E = 2        # Bandwidth in eV given
hc = 1.24e-6       # hc eV·m constant

# Calculate broadening in wavelength
delta_wl = ((wl**2) * delta_E) / hc
# print(delta_lambda)
# Calculate delta
delta_q = ((4 * np.pi * np.sin(np.radians(30)) / (wl**2)) * delta_wl)* 1e-10

print(f"The broadening in q is {delta_q} A")
````

<font color='red'>13. Assume that the resolution of your measurement is $2\theta = 0.2°$, and that you have a perfectly monochromatic beam. For a peak at $2\theta = 10 °$, how large must the nanocrystals be for the calculation to become meaningless?

<font color='red'> **Meaningless is when measured FWHM is larger then resolution of tool.**

<font color='red'> **Known equations:**

<font color='red'> **$ \tau = \frac{K \lambda}{\beta cos \theta} $**


````python
# Constants
wl = 1.54               # X-ray wavelength  Å
K = 0.9                 # Shape factor
peak = 10               # Peak position  degrees
B_in = 0.2              # Instrumental resolution

# Convert
theta = np.radians(peak / 2)
beta = np.radians(B_in)

# Scherrer equation: tau = K * lambda / (beta * cos(theta))
tau = K * wl / (beta * np.cos(theta))

print(f"Maximum nanocrystal size for reliable calculation is {tau} A")
````

### Experimental Geometries & Resolution

Xray Geometry.jpg

The above image defines the experimental geometry of an experiment, where $H$ is the height/width of a square 2D detector, $D$ is the sample-to-detector distance, $2\theta$ is the scattering angle, and $Q$ is the wavevector transfer.

<font color='red'> 14. Using the equation from class, write a user-defined function to convert $2\theta$ to $Q$.

````python
def TwoTheta_to_Q(Twotheta, wl):
    theta_rad = np.radians(Twotheta / 2)  # Convert to theta radians
    Q = (4 * np.pi / wl) * np.sin(theta_rad)
    return Q
````

Q values correspond to real-space distances in the sample by the equation  
<font size=5>$d = \frac{2\pi}{Q}$

<font color='red'> **Known equations:**

<font color='red'> **$ E = \frac{hc }{\lambda} $**

<font color='red'> **$Q = \frac{4π }{\lambda} sinθ$**

<font color='red'> **$Q = \frac{2\pi }{d} $**



<font color='red'> **$\frac{2\pi }{d} = \frac{4π }{\lambda} sinθ$**

<font color='red'> **$\frac{1}{d} = \frac{2 }{\lambda} sinθ$**

<font color='red'> **$\frac{\lambda}{2d} = θ$**

<font color='red'> **$2\theta = \frac{H}{D} $**

<font color='red'> **$\theta = \frac{H}{2D} $**

<font color='red'> **$\frac{\lambda}{2d} = \frac{H}{2D}$**

<font color='red'> **$\frac{\lambda}{Hd} = D$**

````python
# Constants
Height = 10    # Height/width of detector cm
wavelengh = 50   # Wavelengh nm
Energy = 10    # Energy keV
h = 6.626e-34
c = 3e8

# Convert energy to Joules
E = Energy * 1e3 * 1.602e-19
H = Height * 1e-2
d = wavelengh * 1e-9

# Calculate wavelength
lambda_x = h * c / E

# Calculate minimum sample-to-detector distance
D_min = (H * d) / (2 * lambda_x)

print(f"Minimum sample-to-detector distance D is {D_min} m")
````
